```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score,precision_score,recall_score, f1_score,balanced_accuracy_score
```

```python
baseline=pd.read_csv('../Model_Results/01_train_and_test_over_epochs_baseline_model.csv')
gaussian_noise=pd.read_csv("../Model_Results/02_train_and_test_over_epochs_gaussian_noise_model.csv")
gaussian_blur=pd.read_csv("../Model_Results/train_and_test_over_epochs_gaussian_blur_model.csv")
rotate = pd.read_csv("../Model_Results/train_and_test_over_epochs_rotate_model.csv")
color_jitter=pd.read_csv("../Model_Results/train_and_test_over_epochs_cj_c10_model.csv")
```

```python
plt.figure(figsize=(20,10))
plt.plot(baseline['Epochs'],baseline["test_accur"].values, label='Baseline')
plt.plot(color_jitter['Epochs'],color_jitter["test_accur"].values,label='Color Jitter')
plt.plot(rotate['Epochs'],rotate["test_accur"].values, label='Rotate')

plt.plot(gaussian_blur['Epochs'],gaussian_blur["test_accur"].values, label='Gaussian Blur')

plt.plot(gaussian_noise['Epochs'],gaussian_noise["test_accur"].values, label='Gaussian Noise')
plt.xlabel("Epoch")
plt.title("Testing Accuracy vs Epoch")
plt.ylabel("Testing Accuracy")
plt.legend()
plt.show()
```
